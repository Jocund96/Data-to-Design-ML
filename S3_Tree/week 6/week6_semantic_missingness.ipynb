```python
# Import libraries

import pandas as pd
import matplotlib.pyplot as plt
```

# Step 2: Load the Cleaned UHPC Dataset

In this step, we load the cleaned dataset created during Week 5 preprocessing.

This dataset will be used for semantic missingness analysis and further Machine Learning preparation.

```python
# Load the cleaned dataset

df = pd.read_excel("UHPC_Cleaned_Dataset.xlsx")

# Display dataset shape
print("Dataset shape:", df.shape)

# Show first 5 rows
df.head()
```

# Step 3: Check Missing Values

In this step, we identify missing values in the dataset.

Missing values can affect Machine Learning models, so it is important to detect them before preprocessing.

```python
# Check missing values in each column

missing_values = df.isnull().sum()

print(missing_values)
```

# Step 4: Visualize Missing Values

In this step, we create a bar chart to visualize missing values in the dataset.

This helps us identify columns with high amounts of missing data.

```python
# Count missing values

missing_values = df.isnull().sum()

# Create bar chart

plt.figure(figsize=(12, 6))

missing_values.plot(kind='bar')

# Add labels

plt.title("Missing Values in Each Column")
plt.xlabel("Columns")
plt.ylabel("Missing Value Count")

# Rotate column names

plt.xticks(rotation=90)

# Show graph

plt.show()
```

# Step 5: Calculate Missing Value Percentage

In this step, we calculate the percentage of missing values in each column.

This helps us understand how much data is missing relative to the total dataset size.

```python
# Calculate missing value percentage

missing_percentage = (df.isnull().sum() / len(df)) * 100

# Display percentages

print(missing_percentage.sort_values(ascending=False))
```

# Step 6: Identify Highly Missing Features

In this step, we identify features that contain more than 50% missing values.

These features may require special preprocessing, semantic analysis, or removal depending on their importance.

```python
# Find columns with more than 50% missing values

high_missing = missing_percentage[missing_percentage > 50]

print(high_missing)
```
