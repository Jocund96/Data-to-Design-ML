# Step 13 Remove Highly Missing Columns

In this step, we remove columns
containing too many missing values.

This helps:
- improve dataset quality
- simplify preprocessing
- prepare data for machine learning

```python
df_cleaned = df.dropna(
    axis=1,
    thresh=500
)

print(df_cleaned.shape)
```

```python
df_cleaned.head()
```

# Step 16 Fix Dataset Headers

In this step, we assign proper feature names
to the dataframe columns.

This helps:
- remove unnamed headers
- improve dataset readability
- prepare structured data for machine learning

```python
df_cleaned.columns = df_cleaned.iloc[1]

df_cleaned = df_cleaned[2:]

df_cleaned.head()
```

# Step 17  Reset Dataset Index

In this step, we reset dataframe indexing.

This helps:
- create clean row numbering
- improve dataframe readability
- maintain structured preprocessing

```python
df_cleaned = df_cleaned.reset_index(drop=True)

df_cleaned.head()
```

# Step 18  Remove Remaining Extra Header Row

In this step, we remove the remaining unnecessary header row.

This helps:
- clean dataframe structure
- keep only actual data rows
- improve preprocessing quality

```python
df_cleaned = df_cleaned[1:]

df_cleaned = df_cleaned.reset_index(drop=True)

df_cleaned.head()
```

# Step 19  Check Final Dataset Shape

```python
print(df_cleaned.shape)
```

# Step 20 Save Cleaned Dataset

```python
df_cleaned.to_excel(
    "UHPC_Cleaned_Dataset.xlsx",
    index=False
)

print("Dataset saved successfully!")
```
