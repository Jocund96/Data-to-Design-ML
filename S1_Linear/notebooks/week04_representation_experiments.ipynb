# Week 4 — Representation Experiments

## 1. Objective

In Week 4, the goal is to test whether different feature representations improve the Linear Family models.

The experiments include:

- Original UCI/Yeh features
- Yeh/domain engineered ratio features
- Log-transformed skewed features
- More aggressive interaction features

## 2. Why representation matters

Linear models cannot automatically create complex ratios, logarithmic relationships, or interactions unless we provide these features.

For example, water-to-cement ratio is calculated from existing columns, but it gives the model a more useful representation of concrete mix behavior.

## 3. Experiments

The experiments are:

1. Original features
2. Yeh engineered features
3. Log transformed features
4. Aggressive interaction features

## 4. Models

The same Linear Family models are used:

- OLS
- Elastic Net
- Bayesian Ridge
- Polynomial Ridge

## 5. Evaluation

The models are evaluated using:

- MAE
- RMSE
- R
- R²

## 6. Seed test

The best setup is re-run with three different random seeds.

The final result is reported as mean ± standard deviation.
