## Comaprison with non-engineered features (pure dataset)

```python
import pandas as pd

df = pd.read_csv("../Datasets/processed//uci_concrete_dataset/engineered_features.csv")

knn   = pd.read_csv('./results/knn_non-engineered_results.csv')
svr   = pd.read_csv('./results/svr_non-engineered_results.csv')
nusvr = pd.read_csv('./results/nusvr_non-engineered_results.csv')

results = pd.concat([knn, svr, nusvr]).reset_index(drop=True)
print("----PURE DATASET----")
print(results)
```

## Engineered Feature : Water to Cement Ratio
